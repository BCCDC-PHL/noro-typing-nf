```python
import numpy as np
import pandas as pd
import plotly.offline as plo
import plotly.express as plx
import plotly.io as pio
import plotly.graph_objects as pgo
from IPython.display import Markdown as md, HTML, display
from plotly.offline import init_notebook_mode
#import itables 
from itertools import product
import sys, os, re
from glob import glob
import gzip
init_notebook_mode(connected=True)
pio.renderers.default = "notebook"

display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
base = '/home/john.palmer/work/norovirus/results-may-1-comp/filtering/kraken/reports/'
df = pd.read_csv(base+'kraken_combined.csv')
```

```python
ordered_names = df.loc[df['species'] == 'Norovirus'].sort_values('count')['sample']
df["sample"] = pd.Categorical(df["sample"], categories = ordered_names)
df = df.sort_values(by = "sample")
```

```python
fig = plx.bar(df, x='sample',y='count', color='species',width=1600, height=800)
fig.update_xaxes(title_text = "Sample", tickangle=45)
fig.update_yaxes(title_text = "Read Counts")
fig.show()
```
